# Webscrapping NYtimes data

```python
#import library
import requests
from bs4 import BeautifulSoup

#enter rss feed URL
url = "https://rss.nytimes.com/services/xml/rss/nyt/HomePage.xml"

resp = requests.get(url)

soup = BeautifulSoup(resp.content, features="xml")
```

```python
print(soup.prettify())
```

```python
items = soup.findAll('item')
```

```python
print(items)
```

```python
len(items)
```

```python
item = items[1]
```

```python
item
```

```python
#declare empty var
news_items = []
```

```python
#scraping HTML tags: Title, Description, Link and publication date
for item in items:
    news_item = {}
    news_item['title'] = item.title.text
    news_item['description'] = item.description.text
    news_item['link'] = item.link.text
    news_item['pubDate'] = item.pubDate.text
    news_items.append(news_item)
```

```python
#use pandas to create dataframe and to create CSV
import pandas as pd
df = pd.DataFrame(news_items,columns=['title','description','link','pubDate'])
```

```python
df.head()
```

```python
#the csv is created and saved in current working directory
df.to_csv('NYtimes.csv',index=False, encoding = 'utf-8')
```

```python
with open('NY data.txt', mode = 'r', encoding='utf8') as fin:
    cnndata = fin.read()
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

datacounts = word_count(cnndata)
print(datacounts)
```

```python
import operator
sorted_d = sorted(datacounts.items(), key=operator.itemgetter(1))
print(sorted_d)
```
