# Web scrapping BBC RSS fedd data

```python
#import library
import requests
from bs4 import BeautifulSoup

#enter URL
url = "http://rss.cnn.com/rss/cnn_world.rss"

resp = requests.get(url)

soup = BeautifulSoup(resp.content, features="xml")
```

```python
print(soup.prettify())
```

```python
items = soup.findAll('item')
```

```python
print(items)
```

```python
#declare empty var to append data
news_items = []
```

```python
#scarring HTML tags such as Title, Description, Links and Publication date
for item in items:
    news_item = {}
    news_item['title'] = item.title.text
    news_item['description'] = item.description.text
    news_item['link'] = item.link.text
    news_items.append(news_item)
```

```python
print(news_items)
```

```python
news_items[0]
```

```python
#import pandas to create dataframe and CSV
import pandas as pd
df = pd.DataFrame(news_items,columns=['title','description','link','pubDate'])
```

```python
df.head()
```

```python
df.to_csv('CNNdata1.csv',index=False, encoding = 'utf-8')
```

```python
with open('CNNdata1-Copy.txt', mode = 'r') as fin:
    cnndata = fin.read()
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

datacounts = word_count(cnndata)
print(datacounts)
```

```python
import operator
sorted_d = sorted(datacounts.items(), key=operator.itemgetter(1))
print(sorted_d)
```
